```python
from datetime import timedelta, date, datetime
import pandas as pd
import string
import requests
from lxml import html
```

```python
def get_url(date):
    return "https://www.europaplus.ru/index.php?go=Playlist&date={0}&time_start=00.00&time_stop=23.59".format(date)

def daterange(start_date, end_date):
    for n in range(int ((end_date - start_date).days)):
        yield start_date + timedelta(n)
        
def get_page(url):
    page = requests.get(url)
    return (page.status_code, html.fromstring(page.text))

def parse_song(div):
    try:
        at = div[0].text
        if (at != None):
            name = div[2].xpath('strong')[0].text
            if name == None:
                name = div[2].xpath('strong/a')[0].text
            song = div[2].xpath('span')[0].text
            return (at, name, song)
    except:
        return None

def get_songs(page):
    return [parse_song(song) for song in page.xpath('//div[@class="jp_container"]')]

        
```

```python
songs = []
start_date = date(2012, 1, 1)
end_date = date.today()
for single_date in daterange(start_date, end_date):
    url = get_url(single_date.strftime("%Y-%m-%d"))
    print("Parsing url %s" % url)
    (status, page) = get_page(url)
    for song in get_songs(page):
        if song != None:
            full_date = "%s %s" % (single_date.strftime("%Y-%m-%d"), song[0])
            songs.append( (full_date, song[1], song[2]))
```

```python
print("Parsed %s songs" % len(songs))
```

```python
europaplus = pd.DataFrame(songs, columns=['date', 'artist', 'song'])
```
